## Introduction To Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot([1,2,3,4]);
```

```python
x = [1,2,3,4]
y = [11, 22, 33 ,44]
plt.plot(x,y);
plt.show()
```

```python
## method 1
fig = plt.figure() #Creates a figure
ax = fig.add_subplot() #add some axes
plt.show()
```

```python
#Method 2
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x, y)
plt.show()
```

```python
#Method 3 (Recommended)
fig , ax = plt.subplots()
ax.plot(x,y);
type(fig) , type(ax)
```

## Matplotlib example WorkFlow

```python
fig , ax = plt.subplots(figsize = (10,5)) #figsize = (width, height)
ax.plot(x,y)
ax.set(title="Simple Plot" , xlabel="X-axis", ylabel="Y-axis")
#fig.savefig("pathname") if you want to save the figure
plt.show()
```

## Making Figures with Numpy Arrays

```python
# Create some data
x = np.linspace(0,10, 100)
x
```

```python
#Plot the data , create a lineplot
fig , ax = plt.subplots()
ax.plot(x**2)
```

```python
#Scatter Plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x)) #Creates a Scatter plot 
```

```python
fig , ax = plt.subplots()
ax.scatter(x, np.sin(x)) #Creates a sine wave using scatter plot
```

```python
#Make a Plot from dictionary
butter_price = {"Almond":10,
                "Cashew":12,
                "Peanut": 8,}
fig , ax = plt.subplots()
ax.bar(butter_price.keys(),height= butter_price.values()) #BarPlot bar(x, height)
ax.set(title="Butter_Price_DF", xlabel="Nuts" , ylabel="Prices")
```

```python
fig , ax = plt.subplots()
ax.barh(butter_price.keys(),width= butter_price.values()) # Horizontal BarPlot bar(x, width)
ax.set(title="Butter_Price_DF", xlabel="prices" , ylabel="nuts")
```

```python
x = np.random.randn(1000) # Normal standard Distribution
fig , ax = plt.subplots()
ax.hist(x)
```

```python
#subplot option1
fig , ((ax1, ax2) ,(ax3, ax4)) = plt.subplots(nrows=2 , ncols=2, figsize=(10,5))
#Plot to different axes
ax1.plot([1,2,3,4], [3,2,31,2])
ax2.bar(butter_price.keys(), height=butter_price.values())
ax3.scatter(np.random.random(10), np.random.random(10))
ax4.hist(x)
plt.show()
```

```python
#SubPlot Option 2

fig , ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
#Plot to each different axes
ax[0,0].plot(butter_price.keys(), butter_price.values()) ##Accessing axes using index
ax[0,1].scatter(np.random.random(10), np.random.random(10))
#continue

```

```python
heart_disease = pd.read_csv("heart-disease.csv")
```

```python
heart_disease.head()
```

```python
# Create Histogram for Age 
heart_disease["age"].plot.hist(bins=10)
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10, 30) , subplots=True);
```

```python
# Method 2 : 00
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
##Pyplot method
over_50.plot(kind="scatter", x="age", y="chol", c="target") #c -color
```

```python
#OO method
fig , ax = plt.subplots(figsize=(10, 6))    
over_50.plot(kind="scatter", x="age", y="chol", c="target",ax=ax)
ax.set_xlim([45,100])
```

```python
#OO method from scratch
fig ,ax = plt.subplots(figsize=(10,6))
# Plot the data
scatter = ax.scatter(x=over_50["age"], y = over_50["chol"], c=over_50["target"])
#customise the plot
ax.set(title="Heart Disease", xlabel="Age", ylabel="Cholestrol")
#Add Legend
ax.legend(*scatter.legend_elements(), title="target")

#Add Horizontal Line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig , (ax1, ax2)  = plt.subplots(nrows = 2 ,
                                  ncols = 1, figsize = (10,10))
#Add Data to ax1
scatter =ax1.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"]);
#customise ax1
ax1.set(xlabel="age", ylabel="cholestrol", title="Heart Disease(cholestrol)");
#Add Legend
ax1.legend(*scatter.legend_elements(), title="Target");
#Add a Mean Line
ax1.axhline(over_50["chol"].mean() , linestyle="--");

# Add Data to ax2
scatter= ax2.scatter(x=over_50["age"], y=over_50["thalach"],  c=over_50["target"]);
#Customise ax2
ax2.set(xlabel="Age", ylabel="Max Heart Rate", title="Heart Disease (Thalach)");
#Add Legend
ax2.legend(*scatter.legend_elements(), title="Target");
#Add Mean Line for Thalach
ax2.axhline(over_50["thalach"].mean(), linestyle="--");

#Add Title to the figure

fig.suptitle("Heart Disease Analyis", fontsize="20",fontweight="bold", c="Blue" );

```
